## Adjusted cohort

The flowsheet brings in all of the vitals for the jc_uid matched patients from the **draft_cohort**. So we need to bring in the new adjusted_cohort and match these to patient ID and inpatient ID.

```python
# read in the adjusted cohort
demo_df = pd.read_csv("minh/cohort_demo.csv")
demo_df['rit_uid'] = demo_df['jc_uid']
demo_df.head()
```

```python
# match the flowsheet with the cohort
adjusted_flowsheet = flowsheet_df[flowsheet_df.rit_uid.isin(demo_df.jc_uid)]

print("Total rows in adjusted cohort flowsheet WITHOUT time filter: ", adjusted_flowsheet.shape[0])
```

```python
# join the two flowsheets
joined_flowsheet = demo_df.merge(adjusted_flowsheet, on=['rit_uid', 'inpatient_data_id_coded'], how='left')

# change admit_time to datetime
joined_flowsheet['admit_time'] = joined_flowsheet.admit_time + '+00:00'
joined_flowsheet['admit_time'] = pd.to_datetime(joined_flowsheet.admit_time)
joined_flowsheet
```

## Missing patients

After joining the vitals to the adjusted cohort, we want to see how many patients don't have vitals prior to admit. 

From the output, we can see that matching on patient ID and CSN (inpatient ID), we get vitals for all 23,779 unique individuals.

Once we filter this down to vitals that were recorded at or prior to admission (0 hours prior), we have 22,326 patients with vitals and 1,345 without.

The number of patients without vitals increases as the index time is moved earlier.

```python
# filter the adjusted cohort for time (vitals occurring 1 hr prior to admit)

# filter the joined flowsheet+demographics table on time
def filter_time(df, hoursprior):
    flowsheet_time_filtered = df[df.admit_time 
                                 >= df.recorded_time_jittered_utc + timedelta(hours=hoursprior)]
    return flowsheet_time_filtered
    
    
print("COUNTS FOR UNIQUE PATIENTS:")

total_cohort = demo_df.groupby(['rit_uid', 'inpatient_data_id_coded']).ngroups

print("Unique patient in adjusted cohort: ", total_cohort)
print("Matched on patient ID and CSN (no time filter): ", joined_flowsheet.groupby(['rit_uid', 'inpatient_data_id_coded']).ngroups)

for hoursprior in range(5):
    flowsheet_time_filtered = filter_time(joined_flowsheet, hoursprior)
    filtered_cohort = flowsheet_time_filtered.groupby(['rit_uid', 'inpatient_data_id_coded']).ngroups
    missing = total_cohort - filtered_cohort
    print("Filtered for {} hours prior to admit time: {} -- (missing {} patients)".format(hoursprior, filtered_cohort, missing))
```

```python
# split data into train and test so we can look at things separately

# add a year column to make things easier
joined_flowsheet['admit_year'] = pd.DatetimeIndex(joined_flowsheet.admit_time).year
joined_flowsheet['train_set'] = joined_flowsheet['admit_year'] <= 2017

train_flowsheet = joined_flowsheet[joined_flowsheet.train_set == True]
test_flowsheet = joined_flowsheet[joined_flowsheet.train_set == False]

joined_flowsheet
print("Rows of unfiltered training set: ", train_flowsheet.shape[0])
print("Rows of unfiltered test set: ", test_flowsheet.shape[0])
```

```python
# look into the distribution of labels for these different cohorts
def check_labels(df):
    orangebars = []
    bluebars = []

    for hoursprior in range(5):
        df = filter_time(df, hoursprior)
        
        df = df.drop_duplicates(subset=['rit_uid', 'inpatient_data_id_coded'])
        
        count = df.groupby('label').label.count()
        orangebars.append(count[0])
        bluebars.append(count[1])
        print("\n", hoursprior, "hours prior to admit:")
        print(count)
        print("Percent positive labels: ", count[1]/(count[0]+count[1])*100)

    print('\n', orangebars, bluebars)

    df = pd.DataFrame({'orangebars': orangebars, 'bluebars': bluebars})

    # From raw value to percentage
    totals = [i+j for i,j in zip(df['bluebars'], df['orangebars'])]
    bluebars = [i / j * 100 for i,j in zip(df['bluebars'], totals)]
    orangebars = [i / j * 100 for i,j in zip(df['orangebars'], totals)]

    # plot
    barWidth = 0.85
    r = [0,1,2,3,4]
    # Create green Bars
    plt.bar(r, bluebars, color='#b5ffb9', edgecolor='white', width=barWidth, label="Positive case")
    # Create orange Bars
    plt.bar(r, orangebars, bottom=bluebars, color='#f9bc86', edgecolor='white', width=barWidth, label="Negative case")

    # Custom x axis
    plt.xlabel("Hours prior to admit")
    plt.ylabel("Percent")

    # Add a legend
    plt.legend(loc='upper left', bbox_to_anchor=(1,1), ncol=1)

    # Add title
    plt.title("Label distribution across different index times")

    # Show graphic
    plt.show()
```

```python
check_labels(train_flowsheet)
```

## Look into what vitals we have for these cohorts

```python
# check for the types of measures that we find in our cohorts
def check_vitaltype(df):
    for hoursprior in range(5):
        flowsheet_time_filtered = filter_time(df, hoursprior)

        # create bar plot
        counts = flowsheet_time_filtered.groupby('row_disp_name').row_disp_name.count()
        vitals = counts.index
        counts = counts.sort_values(ascending=False)

        plt.bar(counts.index, counts)
        plt.xticks(rotation=90)
        plt.title("Vitals counts for {} hours prior to admit".format(hoursprior))
        plt.show()


    # check out the vitals available without a time filter
    count = joined_flowsheet.groupby('row_disp_name').row_disp_name.count()
    vitals = counts.index
    counts = counts.sort_values(ascending=False)

    plt.bar(counts.index, counts)
    plt.xticks(rotation=90)
    plt.title("Vitals counts for no time filter")
    plt.show()
```

```python
check_vitaltype(train_flowsheet)
```

## Check on vitals per person

Let's see how many vitals each person has.

```python
# print out some summaries for the vital counts per patient event
def event_vital_counts(hoursprior):
    train_filtered = filter_time(train_flowsheet, hoursprior)

    vital_groups = train_filtered.groupby(['row_disp_name', 'jc_uid', 'inpatient_data_id_coded']).label.count().reset_index()
    vital_groups

    # grab the unique vital that we have
    rowdisps = vital_groups.row_disp_name.unique()
    rowdisps

    # print out summary data for each vital's counts
    for disp in rowdisps:
        disp_df = vital_groups[vital_groups.row_disp_name == disp]
        print("\n {} vital counts per patient event, {} hr prior to admit".format(disp, hoursprior))
        print(disp_df.label.describe())

    # get overall stats
    print("\n *** Overall vital counts per patient event, {} hr prior to admit ***".format(hoursprior))
    print(vital_groups.label.describe())
        
        
    # plot the lab counts
    plt.figure(figsize=(10,10))

    ax = sns.boxplot(x='row_disp_name', y='label', data=vital_groups)
    plt.xticks(rotation=90)
    plt.title("Vitals per patient visit {} hr prior to admit".format(hoursprior))
    plt.ylabel("Counts")
    plt.savefig("vitals_boxplot.png")
    
    
    # plot the lab counts
    plt.figure(figsize=(10,10))

    ax = sns.boxplot(x='row_disp_name', y='label', data=vital_groups)
    plt.xticks(rotation=90)
    plt.ylim((0,20))
    plt.title("Vitals per patient visit {} hr prior to admit".format(hoursprior))
    plt.ylabel("Counts")
    plt.savefig("vitals_boxplot_cut.png")
```
