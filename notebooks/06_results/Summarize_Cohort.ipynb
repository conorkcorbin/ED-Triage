## Combine long dataframes

```python
# combine all datasets - vertical stack/concat
full_df_long = pd.concat([vitals, labs, demos], axis=0, join='inner', copy=True)
```

```python
# full_df_long.head()
```

```python
# and un-melt/pivot them
full_df_wide_feats = full_df_long.pivot_table(values='values', 
                                              index='pat_enc_csn_id_coded', 
                                              columns='features')
```

```python
# full_df_wide_feats.head()
```

```python
# LONG --> WIDE
full_df_wide_labels = full_df_long.pivot_table(values='label', 
                                               index='pat_enc_csn_id_coded')
```

```python
full_df_wide = pd.merge(full_df_wide_feats, full_df_wide_labels,
                        on='pat_enc_csn_id_coded', how='inner')
```

### Filter labs for extreme (e.g. 999999) values - pivot first and melt after

```python
# # filter LAB columns with large numbers/feature
# binary_feats = []
# for col in list(full_df_wide_numeric.columns):
    
#     if len(set(full_df_wide_numeric[col].values)) < 3: # binary
#         # categorical type cast
#         full_df_wide_numeric[col] = full_df_wide_numeric[col].astype('category')
#         binary_feats.append(col)
#         print("binary var detected, skipping:", col)
        
#     if col in lab_feats:
#         print("need to filter by quantile:", col)
```

```python
all_feats = set(list(full_df_wide_numeric.columns))
non_lab_feats = all_feats - set(lab_feats)
```

```python
full_df_wide_numeric.shape
```

```python
full_df_wide_labs = full_df_wide_numeric[lab_feats]
full_df_wide_rest = full_df_wide_numeric[list(non_lab_feats)]
```

```python
# filter labs
Q1 = full_df_wide_labs.quantile(0.10)
Q3 = full_df_wide_labs.quantile(0.90)
IQR = Q3 - Q1
full_df_wide_labs_filtered = full_df_wide_labs[~((full_df_wide_labs < (Q1 - 1.5 * IQR)) |(full_df_wide_labs > (Q3 + 1.5 * IQR))).any(axis=1)]
```

```python
full_df_wide_labs_filtered.shape
```

```python
# join with non_labs
full_df_wide_numeric = full_df_wide_labs_filtered.join(full_df_wide_rest, how="inner")
```
