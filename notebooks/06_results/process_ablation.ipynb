```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.metrics import roc_auc_score, average_precision_score
```

```python
full = pd.read_csv("/Users/nostberg/Downloads/results/lightgbm_test_full_results.csv")
meds = pd.read_csv("/Users/nostberg/Downloads/results/lightgbm_test_Meds_results.csv")
imaging = pd.read_csv("/Users/nostberg/Downloads/results/lightgbm_test_Imaging_results.csv")
dx = pd.read_csv("/Users/nostberg/Downloads/results/lightgbm_test_Diagnosis_results.csv")
lab_results = pd.read_csv("/Users/nostberg/Downloads/results/lightgbm_test_Lab_results.csv")
procs = pd.read_csv("/Users/nostberg/Downloads/results/lightgbm_test_Procedures_results.csv")
demo = pd.read_csv("/Users/nostberg/Downloads/results/lightgbm_test_demo_results.csv")
labs_binned = pd.read_csv("/Users/nostberg/Downloads/results/lightgbm_test_labs_results_test_results.csv")
vitals = pd.read_csv("/Users/nostberg/Downloads/results/lightgbm_test_vitals_test_results.csv")
```

```python
def bootstrap(df):
    aucs = []
    prc = []
    for i in range(10000):
        sample = df.sample(n=1000)
        aucs.append(roc_auc_score(sample["label"], sample["predictions"]))
        prc.append(average_precision_score(sample["label"], sample["predictions"]))
    
    return aucs, prc
```

```python
results = pd.DataFrame(columns=['AUROC value','AUROC 95% CI','AUPRC value','AUPRC 95% CI'])

for df in [full, meds, imaging, dx, lab_results, procs, demo, labs_binned, vitals]:
    aucs, prc = bootstrap(df)
    results = results.append({
     "AUROC value": roc_auc_score(df["label"], df["predictions"]),
     "AUROC 95% CI":  (np.percentile(np.array(aucs), 2.5), np.percentile(np.array(aucs), 97.5)),
     "AUPRC value": average_precision_score(df["label"], df["predictions"]),
     'AUPRC 95% CI': (np.percentile(np.array(prc), 2.5), np.percentile(np.array(prc), 97.5))
      }, ignore_index=True)
```

```python
results.rename(index={0:"Full", 1:"Meds", 2:"Imaging", 3:"Dx", 
                      4:"Labs", 5:"Procedures", 6:"Demographics", 7:"Lab_results", 8:"vitals"})
```
